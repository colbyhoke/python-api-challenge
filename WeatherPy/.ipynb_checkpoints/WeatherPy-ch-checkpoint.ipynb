```python
# Colby Alexander Hoke
# UNC Data Analytics Bootcamp, July 2020
# CC-BY-SA
# --------------------------------------
# INPUT: No input
# DATA: Generates random numbers for latitude and longitude
#       Then uses citipy to find nearest city to generated coordinates
#       Then pulls weather data from cities
#
# RETURN: Several analyses, detailed below
# --------------------------------------

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress

# Turn on to seed the random to make it reproducable for testing
#np.random.seed(42)

# Import API key
from config import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Build a dataframe with everything we want
city_df = pd.DataFrame(columns = ['City','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed','Country','Date'])

# Populate 'City' series with list of city names
city_df['City'] = cities 

# API URL and parameters
base_url = 'http://api.openweathermap.org/data/2.5/weather?q='
params = {'units':'imperial', 'appid': weather_api_key}

print('-------------------------------')
print('****Begin processing city data****')
print('-------------------------------')

# Loop through all cities in the dataframe, gather weather data, and store in dataframe
for index, row in city_df.iterrows():
    city = row['City']
    
    # Call the API and convert to imperial units
    response = requests.get(base_url + city + "&units=imperial", params=params).json()
    
    print(f'Processing record: {index} | {city}.')
    
    #Put city data into a dataframe
    try:   
        city_df.loc[index, 'Lat'] = response['coord']['lat']
        city_df.loc[index, 'Lng'] = response['coord']['lon']
        city_df.loc[index, 'Max Temp'] = response['main']['temp_max']
        city_df.loc[index, 'Humidity'] = response['main']['humidity']
        city_df.loc[index, 'Cloudiness'] = response['clouds']['all']
        city_df.loc[index, 'Wind Speed'] = response['wind']['speed']
        city_df.loc[index, 'Country'] = response['sys']['country']
        city_df.loc[index, 'Date'] = response['dt']
    
    # Catch exceptions, skip them, and drop them from the dataframe
    except (KeyError, IndexError):
        print('City not found. Skipping and dropping.')
        
        # Drop the cities that aren't found
        city_df.drop([index], inplace=True)
        
    print("------------")    
print('-------------------------------')
print('****Data processing complete****')
print('-------------------------------')
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
```

```python
# Create Northern and Southern Hemisphere DataFrames
n_hemi_df = clean_city_df.loc[clean_city_df['Lat']>=0]
s_hemi_df = clean_city_df.loc[clean_city_df['Lat']<0]

# --------------------------------------
# Function to calculate and plot linear regressions
# 
# Takes in:
#     hemisphere: 'North' or 'South'
#     dimension: 'Max Temp','Humidity','Cloudiness', or 'Wind Speed'
#     anno_x: x value for where you want the line equation annotated
#     anno_y: y value for where you want the line equation annotated
#     *Tweak anno_x and anno_y per chart for more control over line annotation display
# --------------------------------------
 
def linreg(hemisphere, dimension):
    
    # If hemisphere is North
    if hemisphere == 'North':
        x_ax = n_hemi_df['Lat']
        y_ax = n_hemi_df[dimension]
    
    elif hemisphere == 'South':
        x_ax = s_hemi_df['Lat']
        y_ax = s_hemi_df[dimension]
      
    fig = plt.figure()
    plt.scatter(x=x_ax, y=y_ax, marker='o', c='lightblue', edgecolors='black', s=30)

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_ax.astype(float), y_ax.astype(float))

    regress_vals = x_ax * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
        
    plt.plot(x_ax,regress_vals,"r-")
        
    # Add labels
    plt.xlabel(f'Latitude \n (Line equation: {line_eq})')
        
    if dimension == 'Max Temp':
        title = f'{hemisphere}ern Hemisphere Latitude vs. Temperature'
        plt.title(f'{title} ({date})')
        plt.ylabel(f'{dimension} (Fahrenheit)')
    elif dimension == 'Humidity':
        title = f'{hemisphere}ern Hemisphere Latitude vs. Humidity'
        plt.title(f'{title} ({date})')
        plt.ylabel(f'{dimension} (%)')
    elif dimension == 'Cloudiness':
        title = f'{hemisphere}ern Hemisphere Latitude vs. Cloudiness'
        plt.title(f'{title} ({date})')
        plt.ylabel(f'{dimension} (%)')
    elif dimension == 'Wind Speed':
        title = f'{hemisphere}ern Hemisphere Latitude vs. Wind Speed'
        plt.title(f'{title} ({date})')
        plt.ylabel(f'{dimension} (mph)')
        
    plt.show()
    fig.savefig(f'./output_data/{title}.png',bbox_inches='tight')
    
    return
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Call linreg function with variables
linreg('North','Max Temp')
```
